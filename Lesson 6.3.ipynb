```python
data = pd.read_csv("penguins.csv")
data.head()
```

## Scatter Plot

```python
plt.scatter(data['Flipper Length (mm)'], data['Body Mass (g)'])
plt.show()
```

```python
colors = {'Adelie Penguin (Pygoscelis adeliae)':'red', 'Chinstrap penguin (Pygoscelis antarctica)':'green', 'Gentoo penguin (Pygoscelis papua)':'blue'}
plt.scatter(data['Flipper Length (mm)'], data['Body Mass (g)'], color=data['Species'].map(colors))
plt.show()
```

## Histogram

```python
# the histogram of the data (6 is numbe rof bins)
plt.hist(data['Flipper Length (mm)'],15, facecolor='green')

plt.xlabel('Length(mm)')
plt.ylabel('Numbe rof Penguins')
plt.title("Histogram of Flipper Length")
plt.grid(True)

plt.show()
```

```python
# the histogram of the data (6 is numbe rof bins)
plt.hist([Female['Flipper Length (mm)'],Male['Flipper Length (mm)']],3,density=True, color=['red','blue'], label=['Female','Male'])
plt.legend(prop={'size': 10})
plt.xlabel('Length(mm)')
plt.ylabel('Probability')
plt.title("Histogram of Flipper Length")
plt.grid(True)

plt.show()
```

## Box Plot

```python
## https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.boxplot.html
plt.boxplot(data['Flipper Length (mm)'])
plt.title("Flipper Length")
plt.ylabel("Length (mm)")
plt.show()
```

```python
Female = data.loc[data['Sex'] == 'FEMALE']
Male = data.loc[data['Sex'] == 'MALE']

plt.title('Flipper Length')
plt.boxplot([Female['Flipper Length (mm)'],Male['Flipper Length (mm)']])
plt.ylabel("Length (mm)")
plt.xticks([1,2],["F","M"])
plt.show()
```

## Plotting Geographic Data

```python
court_houses = pd.read_csv('https://files.ontario.ca/opendata/latitude_and_longitude_of_ontario_courthouses.csv')
court_houses.head()
```

```python
print(court_houses[["'Bldg_Longitude'"]].min())
print(court_houses[["'Bldg_Longitude'"]].max())     
print(court_houses[["'Bldg_Latitude'"]].min())
print(court_houses[["'Bldg_Latitude'"]].max())

BBox = (court_houses.iloc[:,4].min(),court_houses.iloc[:,4].max(),court_houses.iloc[:,3].min(), court_houses.iloc[:,3].max())
print(BBox)
```

https://www.openstreetmap.org/export

```python
ontario = plt.imread("map.png")

fig, ax = plt.subplots(figsize = (8,7))
ax.scatter(court_houses.iloc[:,4]+2, court_houses.iloc[:,3]+2, alpha= 0.4, c='r', s=20)
ax.set_title('Plotting Court Houses')
ax.set_xlim(BBox[0],BBox[1])
ax.set_ylim(BBox[2],BBox[3])
ax.imshow(ontario, zorder=0, extent = BBox, aspect= 'equal')
```
